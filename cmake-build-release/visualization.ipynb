```python
import matplotlib.pyplot as plt

def read_data(filename):
    with open(filename, 'r') as file:
        data = []
        for line in file:
            for num in line.split():
                try:
                    data.append(float(num))
                except ValueError:
                    continue
        return data
```

```python
# Read data from three files
tour_lengths1 = read_data('strategy1_tour_lengths_100.txt')
tour_lengths2 = read_data('strategy2_tour_lengths_100.txt')
# tour_lengths3 = read_data('sa_current_tour_lengths_100.txt')
# tour_lengths4 = read_data('sa_best_tour_lengths_100.txt')

# Create a figure and a set of subplots
fig, ax = plt.subplots()

# Plot data from the three files
ax.plot(tour_lengths1, label='Best Improvement')
ax.plot(tour_lengths2, label='First Improvement')
# ax.plot(tour_lengths3, label='SA (Current)')
# ax.plot(tour_lengths4, label='SA (Best)', linestyle=':', color='lightgray')

# Set labels
ax.set_xlabel('Iteration')
ax.set_ylabel('Tour Length')
ax.set_title('Tour Lengths Over Iterations (n=100)')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```
